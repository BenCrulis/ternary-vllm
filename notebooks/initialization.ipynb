```python
import numpy as np
import matplotlib.pyplot as plt

eps = 5e-2
```

```python
np.random.seed(2222)

# generate random points
x = np.random.normal(0.0, 2, 30)
# x = np.random.standard_cauchy(40)

x = np.clip(x, -5, 5)

lim = 5.1

horizontal = np.zeros_like(x) + np.random.rand(*x.shape)*eps

# plot the points vertically
plt.scatter(horizontal, x)
plt.xlim(-1, 1)
plt.ylim(-lim, lim)
```

```python
x_abs = np.abs(x)

# plot the points vertically

plt.scatter(horizontal, x_abs)
plt.xlim(-1, 1)
plt.ylim(-lim, lim)

print(x_abs.max())
```

```python
# assign labels to the point

y = np.zeros_like(x)
y[x_abs > x_abs.max() / 2] = 1

# plot the points vertically
plt.scatter(horizontal, x_abs, c=y)
plt.xlim(-1, 1)
plt.ylim(-lim, lim)
# plot the centroids as a x
centroid = x_abs.max()
plt.scatter(0, 0.0, marker='x', s=100, c='blue')
plt.scatter(0, centroid, marker='x', s=100, c='orange')
```

```python
centroid = x_abs[y == 1].mean()

# iterate the k-means algorithm

l = np.square(x_abs[y == 1] - centroid).sum() + np.square(x_abs[y == 0] - 0).sum()

for i in range(10):
    print("i=", i)
    print("l=", l)
    # assign each point to the closest centroid
    y = np.abs(x_abs - centroid) < np.abs(x_abs)
    # update the centroid
    centroid = x_abs[y == 1].mean()
    print(centroid)
    l = np.square(x_abs[y == 1] - centroid).sum() + np.square(x_abs[y == 0] - 0).sum()

    # plot the points vertically
    plt.scatter(horizontal, x_abs, c=y)
    plt.xlim(-1, 1)
    plt.ylim(-lim, lim)
    plt.scatter(0, 0.0, marker='x', s=100, c='blue')
    plt.scatter(0, centroid, marker='x', s=100, c='orange')
    plt.show()
```

```python
# perform cluster assignment in original space
y = np.zeros_like(x)
y[x > centroid/2.0] = 1
y[x < -centroid/2.0] = 2

# plot the points vertically
plt.scatter(horizontal, x, c=y)
plt.xlim(-1, 1)
plt.ylim(-lim, lim)
```
